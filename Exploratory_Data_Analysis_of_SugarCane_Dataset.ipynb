```python
import re
import numpy as np

sug_cane_df['Acreage (Hectare)'] = sug_cane_df['Acreage (Hectare)'].astype(str)
sug_cane_df['Acreage (Hectare)'] = sug_cane_df['Acreage (Hectare)'].apply(lambda x: re.sub(r'\D', '', x))
sug_cane_df['Acreage (Hectare)'] = sug_cane_df['Acreage (Hectare)'].replace('', np.nan).astype(float)
mean = sug_cane_df['Acreage (Hectare)'].mean()
```

```python
mean
```

```python
sug_cane_df['Acreage (Hectare)'].fillna(mean,inplace=True)
```

```python
sug_cane_df.loc[sug_cane_df.loc[:,'Yield (Kg / Hectare)'].isna(),:]
```

```python
import re

sug_cane_df['Yield (Kg / Hectare)'] = sug_cane_df['Yield (Kg / Hectare)'].str.replace(r'\D', '', regex=True)
sug_cane_df['Yield (Kg / Hectare)'] = sug_cane_df['Yield (Kg / Hectare)'].astype(float)
mean = sug_cane_df['Yield (Kg / Hectare)'].mean()
```

```python
sug_cane_df['Yield (Kg / Hectare)'].fillna(mean,inplace=True)
```

```python
sug_cane_df.tail()
```

```python
sug_cane_df['Production (Tons)'] = sug_cane_df['Production (Tons)'].str.replace('.', '').astype(float)
sug_cane_df['Production per Person (Kg)'] = sug_cane_df['Production per Person (Kg)'].str.replace(',', '').astype(float)
```

**Production Comparison: We can compare the production of different countries or continents to identify the top producers. This can be done by grouping the data by country or continent and calculating the total production.**

```python
sug_cane_df.sort_values('Production (Tons)',ascending=False).head()
```

```python
sug_cane_df.sort_values('Production (Tons)',ascending=True).head(5)
```

```python
sug_cane_df.groupby(['Continent']).sum().sort_values(by='Production (Tons)',ascending=False)
```

```python
sug_cane_df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
import matplotlib.pyplot as plt
```

```python
x=sug_cane_df.sort_values('Production (Tons)',ascending=False).head()['Country']
y=sug_cane_df.sort_values('Production (Tons)',ascending=False).head()['Production (Tons)']
```

```python
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y)

# Customize the plot
ax.set_xlabel('COUNTRIES')
ax.set_ylabel('PRODUCTION')
ax.set_title('TOP COUNTRIES PRODUCTION')

# Show the plot
plt.show()
```

```python
x=sug_cane_df.groupby('Continent').agg({'Production (Tons)':'sum'}).sort_values(by='Production (Tons)',ascending=False).index[0:]
y=sug_cane_df.groupby('Continent').agg({'Production (Tons)':'sum'}).sort_values(by='Production (Tons)',ascending=False)['Production (Tons)'].values
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y)

# Customize the plot
ax.set_xlabel('CONTINENTS')
ax.set_ylabel('PRODUCTION')
ax.set_title('TOP CONTINENT PRODUCTION')

# Show the plot
plt.show()
```

**Acreage Analysis: You can analyze the acreage (hectares) to identify countries or continents with large or small agricultural land holdings. This information can be used to assess the availability of land for farming and its impact on production.**

```python
sug_cane_df.columns
```

```python
sug_cane_df.sort_values('Acreage (Hectare)',ascending=False).head()
```

```python
sug_cane_df.sort_values('Acreage (Hectare)',ascending=True).head()
```

```python
x=sug_cane_df.sort_values('Acreage (Hectare)',ascending=False).head()['Country']
y=sug_cane_df.sort_values('Acreage (Hectare)',ascending=False).head()['Acreage (Hectare)']
# Create a figure and axis
fig, ax = plt.subplots()

# Create the pie chart
ax.pie(y, labels=x, autopct='%1.2f%%')

# Customize the plot
ax.set_title('Sugarcane Acreage (Hectare) by Country')

# Equal aspect ratio ensures that pie is drawn as a circle
ax.axis('equal')

# Show the plot
plt.show()
```

```python
explode = [0.1, 0, 0, 0, 0]  # To explode the first slice

# Create a figure and axis
fig, ax = plt.subplots()

# Create the pie chart
colors = ['#FF9999', '#FFCC99', '#FFFF99', '#99FF99', '#99CCFF', '#CCCCCC']
ax.pie(y, labels=x, autopct='%1.1f%%', explode=explode, colors=colors, startangle=90, shadow=True)

# Customize the plot
ax.set_title('Sugarcane Production by Continent')
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

# Legend
legend_labels = ['{} - {:.2f} Tons'.format(x, prod) for x, prod in zip(x,y)]
ax.legend(legend_labels, loc='best', bbox_to_anchor=(1, 0.5))

# Show the plot
plt.show()
```
